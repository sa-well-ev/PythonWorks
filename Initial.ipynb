```python
import pandas as pd

file_path = "./data/Transactions.csv"

if file_path:
    v_df1 = pd.read_csv(file_path, sep=';', encoding='utf-8')    
else:
    print("Файл не выбран.")

v_df1 = v_df1[['Плата', 'Дата', 'Сумма', 'Категория2', 'Источник']]
v_df1.columns=['payment', 'date', 'amount', 'category', 'source']
# Преобразование столбца amount в формат float
v_df1.amount = (
    v_df1.amount
    .str.replace('\s+', '', regex=True)
    .str.replace(',', '.')
    .astype(float)
)
# Преобразование столбца date в формат datetime
v_df1.date = pd.to_datetime(v_df1.date, format='%d.%m.%Y %H:%M')

# Фильтрация данных по payment
v_payments = v_df1.loc[v_df1.payment == 1]
v_incomes = v_df1.loc[v_df1.payment == 0]

# Группировка по дате и категории с подсчетом суммы
v_PmonthMean = (
    v_payments
    .groupby(['category', v_df1.date.dt.to_period('M')])
    .amount
    .mean()
    )

# Группировка по дате и категории с подсчетом суммы
v_ImonthMean = (
    v_incomes
    .groupby(['category', v_df1.date.dt.to_period('M')])
    .amount
    .sum()
    )


v_df1
# v_df1.info()
# v_PmonthMean
# v_ImonthMean
# v_ImonthMean.loc['Зарплата Д']
# v_ImonthMean.index.get_level_values('category').unique()
```
